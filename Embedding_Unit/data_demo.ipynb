```python
import pandas as pd
df1=pd.read_csv('/home/g6/Desktop/Fawzy/ag_news_csv/AG_News.csv')
df2 = pd.read_csv("/home/g6/Desktop/Fawzy/multinews_combined.csv")
df1.head()
```

```python
df1.drop(["ID", "label"], axis=1, inplace=True)
df1.head()
```

```python
df2.drop(["Id", "Unnamed: 0"], axis=1, inplace=True)
```

```python
df2.head()
```

```python
mnews_df = df2
agnews_df = df1
```

```python
# title is the first 10 words of he summary column
mnews_df["title"] = mnews_df["summary"].apply(lambda x: " ".join(x.split()[:15]))
mnews_df.head()
```

```python
# Rename description column to document
agnews_df.rename(columns={"description": "summary"},inplace=True)
agnews_df["document"] = agnews_df["title"] + "\n" + agnews_df["summary"]
```

```python
import h5py
with h5py.File('AG_News.hdf5', 'r') as f:
     data=f['train']
     print(data.shape)
     #add a column to the dataframe with the embeddings
     agnews_df['embeddings'] = list(data)
```

```python
agnews_df.drop_duplicates(subset=['title', 'summary'], inplace=True)
```

```python
with h5py.File('Multinews.hdf5', 'r') as f:
     data=f['train']
     print(data.shape)
     #add a column to the dataframe with the embeddings
     mnews_df['embeddings'] = list(data)
```

```python
merged_df = pd.concat([agnews_df, mnews_df], ignore_index=True)
```

```python
import numpy as np
mergedData = merged_df['embeddings'].values
```

```python
#unpack numpy array mergedData
mergedData2 = np.array([x for x in mergedData])
```
